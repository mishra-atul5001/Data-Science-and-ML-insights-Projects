```python
#Let us Check how many Students totally passed in All Subjects:
df['OverAllPassingStatus'] = df.apply(lambda x: 'Failed!' if x['MathPassingStatus'] == 'Failed!' or x['WritingPassingStatus'] == 'Failed!' or x['ReadingPassStatus'] == 'Failed!'
                                      else 'Passed!' , axis =1)
df.OverAllPassingStatus.value_counts()
```

```python
#Finding % of Marks:
df['Total_Marks'] = df['math score'] + df['reading score'] + df['writing score']
df['Percent'] = df['Total_Marks']/3
```

```python
p = sns.countplot(x="Percent" , data = df , palette = "muted")
_ = plt.setp(p.get_xticklabels(),rotation = 0)
```

```python
#Let us do the grading for the students now:
def GetGrade(Percent,OverAllPassingStatus):
    if(OverAllPassingStatus == 'Failed!'):
        return 'Failed'
    if(Percent >= 80):
        return 'A'
    if(Percent >= 70):
        return 'B'
    if(Percent >= 60):
        return 'C'
    if(Percent >= 50):
        return 'D'
    if(Percent >= 40):
        return 'E'
    else:
        return 'Failed!'
    
```

```python
df['Grade'] = df.apply(lambda x: GetGrade(x['Percent'], x['OverAllPassingStatus']),axis =1)
df.Grade.value_counts()
```

```python
#Plotting Grades in an Obtained Order
sns.countplot(x="Grade" , data=df,order = ['A','B','C','D','E','F'] , palette = "muted")
plt.show()
```

```python
#Comparing All 3 Scores with a Histogram Plot..!!
fig , axes = plt.subplots(1,3,sharey=True,figsize=(18,5))
ax1,ax2,ax3 = axes.flatten()
ax1.hist(df['math score'] , bins=10,color='red')
ax2.hist(df['reading score'],bins=10,color='blue')
ax3.hist(df['writing score'],bins=10,color='orange')
ax1.set_xlabel('MATH',fontsize='large')
ax1.set_ylabel('SCORE',fontsize='large')
ax2.set_xlabel('READING',fontsize='large')
ax3.set_xlabel('WRITING',fontsize='large')
plt.suptitle('Score Comparison',ha = 'center',fontsize='x-large')
plt.show()
```

```python
#KDE PLOT showing all Subjects Variation
import warnings
warnings.filterwarnings('ignore')

sns.kdeplot(df['math score'],shade=True,color='red',alpha=0.9)
sns.kdeplot(df['reading score'],shade = True,color='blue',alpha=0.6)
sns.kdeplot(df['writing score'],shade=True,color='orange',alpha=0.3)
plt.show()
```

```python
#Let us Make a New DATA FRAME for Acceptance of these Marks satisfying for a Admission College or not.
df['acceptance_denied'] = np.where(df['Percent'] >=70,'Accepted','Denied')
df.acceptance_denied.value_counts()
```

```python
#Let's plot this Variation with GENDER as a Constraint to see how many MALE or FEMALE Students have been accpeted with respected Scores
math_facet_grid = sns.FacetGrid(df,col='acceptance_denied',hue='gender')
math_facet_grid.map(sns.kdeplot,'math score')
plt.legend()
plt.show()
```

```python
#Similarly we Can Look for Reading Score and Writing Score..!!
reading_facet_grid = sns.FacetGrid(df,col='acceptance_denied',hue='gender')
reading_facet_grid.map(sns.kdeplot,'reading score')
writing_facet_grid = sns.FacetGrid(df,col='acceptance_denied',hue='gender')
writing_facet_grid.map(sns.kdeplot,'writing score')

plt.legend()
plt.show()
```

```python
#Combining to all the Above...We can draw a Pairplot

# A Pair Plot to see the Affect of Each Score with the Other Subject Score.
#We find a 3x3 matrix as 3 Subjects Variation with respect to each subject is plotted.
g=sns.pairplot(df.iloc[:,[0,5,6,7]],hue = 'gender',diag_kind='kde')
plt.show()
```

```python
#We can finally even break down how many male Students got Accepted along with Female Students and How many got Denied..!!
df.groupby(['acceptance_denied','gender']).OverAllPassingStatus.value_counts().unstack()
```
